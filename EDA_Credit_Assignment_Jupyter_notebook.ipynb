## 1. Handling Missing Values

#### Finding Null values (normalized)

```python
df_applications.isna().sum().sort_values(ascending=False)*100/len(df_applications)
```

#### Iterating through each column in dataset and appending a list of above 40% null values

```python
counter = 0
col_list = []
for col in df_applications:
    counter += 1
    col_arr = df_applications[col]
    null_counts = col_arr.isna().sum()
    null_pct = round(null_counts/len(df_applications)*100,2)
    if null_pct>40:
        print(f"Column ({counter}): {col}, NullCounts: {null_counts}, Null%: {null_pct}")
        col_list.append(col)
```

#### Dropping columns as these have high missing values above 40% of dataset, so will not cover in analysis

```python
df_applications.drop(col_list,axis=1,inplace=True)
```

#### There are some missing values in CODE_GENDER as 'XNA', we may replace these values with 'F' as the mode of column is 'F'

```python
df_applications.CODE_GENDER.replace('XNA',np.NaN,inplace=True)
df_applications.CODE_GENDER.fillna('F',inplace=True)
```

#### occupation type has missing values, we may impute these with unknown category

```python
df_applications.OCCUPATION_TYPE= df_applications.OCCUPATION_TYPE.fillna('Unknown')
```

```python
df_applications.OCCUPATION_TYPE.isna().sum()
```

## 2. Data Standardisation

#### Target column has two type of value 1 represents who had defaulted and 0 represented who had not defaulted
#### A. Replacing the 0 and 1 with category Defaulter and Regular

```python
df_applications.TARGET = df_applications.TARGET.apply(lambda x: 'Regular' if x == 0 else 'Defaulter')
```

```python
df_applications.TARGET.head()
```

```python
df_applications.FLAG_OWN_CAR = df_applications.FLAG_OWN_CAR.apply(lambda x : 'Yes' if x=='Y' else 'No')
df_applications.FLAG_OWN_REALTY = df_applications.FLAG_OWN_REALTY.apply(lambda x : 'Yes' if x=='Y' else 'No')
```

#### B. Converting negative days to positive

```python
df_applications.DAYS_BIRTH *= -1
```

#### C. Creating a feature to have birth days as years for better readability

```python
df_applications['AGE_YEARS'] = df_applications.DAYS_BIRTH.apply(lambda days:float(days/365))
```

```python
def df_to_percentage(df):
    for col in df:
        df[col] = df[col].apply(lambda x:(x/df[col].sum())*100)
    return df
```
